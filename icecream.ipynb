```python
import pandas as pd
import matplotlib.pyplot as plt
```

Lê os dados da tabela

```python
icecr = pd.read_csv('icecream.csv')
icecr
```

Traz a tabela em ordem decrescente

```python
icecr = icecr.sort_values('Temperature', ascending = False)
icecr
```

Transforma a tabela de temperatura fahrenheit para celsius fazendo a conversão

```python
celsius = (icecr['Temperature'] - 32) * 5/9
icecr['Temperature'] = celsius

icecr
```

Tranforma os dados apresentados em um gráfico e exporta para pdf

```python
plt.title('Sorteves vendidos de acordo com a temperatura')

plt.plot(icecr['Temperature'], icecr['Sales'], 'r.--', label = 'US$')

plt.xlabel('Temperatura (Celsius)')
plt.ylabel('Preço sorvete')
plt.legend

plt.savefig('preco_sorvete_por_temperatura.pdf')
plt.show()


```
